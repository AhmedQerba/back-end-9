##Data Cleaning

```python
df.sample(5)
```

Cleaning values of Engine_cylinder_number and Engine_total_volume columns

```python
df['Engine_cylinder_number'] = df['Engine_cylinder_number'].str.replace('cyl', '').str.replace('0 L', '').str.replace('2 L', '').str.replace('-', '')
df['Engine_total_volume'] = df['Engine_total_volume'].str.replace('L', '')
df = df.replace('', np.nan)
df['Engine_cylinder_number'] = df['Engine_cylinder_number'].astype(float)
df['Engine_total_volume'] = df['Engine_total_volume'].astype(float)
```

Cleaning values of fuel_comsumption_liter & fuel_comsumption_km columns

```python
df['fuel_comsumption_liter'] = df['fuel_comsumption_liter'].str.replace('L', '')
df['fuel_comsumption_km'] = df['fuel_comsumption_km'].str.replace('km', '')
df['fuel_comsumption_liter'] = df['fuel_comsumption_liter'].str.replace('-', '')
df = df.replace('', np.nan)
df['fuel_comsumption_liter'] = df['fuel_comsumption_liter'].astype(float)
df['fuel_comsumption_km'] = df['fuel_comsumption_km'].astype(float)
```

Cleaning values of Doors column

```python
df['Doors'] = df['Doors'].str.replace('Doors', '').str.replace('12 Seats', '6').str.replace('3 Seats', '2').str.replace('6 Seats', '4').str.replace('9 Seats', '6').str.replace('8 Seats', '4').str.replace('2 Seats', '2').str.replace('4 Seats', '4').str.replace('5 Seats', '3').str.replace('7 Seats', '4').str.replace('Seats', '')
df['Doors'] = df['Doors'].astype(float)
```

Cleaning values of Seats column

```python
df['Seats'] = df['Seats'].str.replace('Seats', '')
df['Seats'] = df['Seats'].astype(float)
```

Cleaning values of body column

```python
df['BodyType'] = df['BodyType'].str.replace('/', '')
```

Cleaning values of carColor colum

```python
df['carColor'] = df['carColor'].str.replace('-', '')
```

Cleaning values of Kilometers column

```python
df['Kilometres'] = df['Kilometres'].str.replace('-', '').str.replace('/', '').str.replace(' ', '')
df['Kilometres'].replace('', np.nan, inplace=True)
df['Kilometres'] = df['Kilometres'].astype(float)
```

Cleaning Price column

```python
df['Price'].replace('POA','', inplace=True)
df = df.replace('', np.nan)
df['Price'] = df['Price'].astype(float)
```

Cleaning Transmission column

```python
df['Transmission'].replace('-','', inplace=True)
df = df.replace('', np.nan)
df['Transmission'].describe()
```

Cleaning values of Year column

```python
df['Year'] = df['Year'].astype(str).str[:4]
df = df.replace('nan', np.nan)
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
```

Separating labels

```python
X=df.drop('Price', axis=1)
Y=df['Price']
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler
target_col_name='Price'
col_for_scaled=[]
all_col_nm= X.select_dtypes(include=['number']).columns.tolist()

for i in range(len(all_col_nm)):
    xx=list(X[all_col_nm[i]].unique())

    if max(xx)>2:
        if all_col_nm[i]==target_col_name:
            pass
        else:
            col_for_scaled.append(all_col_nm[i])
    else:
        pass

x_cols_for_scale=X[col_for_scaled]
scaler = StandardScaler()
scaler.fit(x_cols_for_scale)


x_transform_dt = scaler.transform(x_cols_for_scale)
x_scaled_df = pd.DataFrame(x_transform_dt, columns=col_for_scaled)
X.drop(col_for_scaled,axis=1,inplace=True)
X.reset_index(drop=True, inplace=True)
x_scaled_df.reset_index(drop=True, inplace=True)
X=pd.concat([X, x_scaled_df], axis=1)
X.head()
```

##Encoding

Perform Lable encoding

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
encoding=LabelEncoder()
X["Transmission"]=encoding.fit_transform(X["Transmission"])
X["UsedOrNew"]=encoding.fit_transform(X["UsedOrNew"])
```

Performing one-hot encoding

```python
numerical_cols = X.select_dtypes(include=['int', 'float']).columns.tolist()
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()

encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')

# Fit and transform the categorical columns
encoded_categorical = encoder.fit_transform(X[categorical_cols])

# Create DataFrame with the encoded categorical columns
encoded_df = pd.DataFrame(encoded_categorical, columns=encoder.get_feature_names_out(categorical_cols))

# Combine encoded categorical columns with numerical columns
X = pd.concat([X[numerical_cols], encoded_df], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2, random_state=1232224)
```

```python
from sklearn.ensemble import RandomForestRegressor as randomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
rfr=randomForestRegressor()
rfr.fit(x_train,y_train)
rfrtd=rfr.predict(x_test)
rfrr2=r2_score(y_test,rfrtd)
rfrr2
```

Model Evaluation

```python
# make predictions on the testing set
y_pred = rfr.predict(x_test)

# Calculate MSE
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)

# Calculate RMSE
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)

# Calculate MAE
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (MAE):", mae)

# Calculate R2 score
r2 = r2_score(y_test, y_pred)
print("R-squared (R2) Score:", r2)
```

##Clustering

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Properly import Plotly's graph_objects
import plotly.graph_objects as go  # Import graph_objects and alias as 'go'


# Assuming 'X' is your combined DataFrame with numerical and encoded categorical columns
# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Standardize the data

# Determine the optimal number of clusters with the elbow method
wcss = []
max_clusters = 10  # Test different cluster counts from 1 to 10
for i in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)  # Ensure you're using the correct variable
    wcss.append(kmeans.inertia_)

# Create a scatter plot with lines and markers
fig = go.Figure(
    go.Scatter(
        x=list(range(1, max_clusters + 1)),  # x-axis: number of clusters
        y=wcss,  # y-axis: Within-Cluster Sum of Squares
        mode='lines+markers',  # Line plot with markers
        marker=dict(size=8, color='blue'),  # Marker size and color
        line=dict(width=2),  # Line width
    )
)

# Customize the plot layout
fig.update_layout(
    title="Elbow Method to Determine Optimal Clusters",
    xaxis=dict(title="Number of Clusters"),
    yaxis=dict(title="Within-Cluster Sum of Squares (WCSS)"),
    width=800,  # Size of the plot (equivalent to figsize)
    height=600,
)

# Display the Plotly plot
fig.show()

```

```python
import plotly.express as px
import plotly.graph_objects as go
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming 'X_scaled' is the scaled data and 'optimal_clusters' is defined
optimal_clusters = 5  # Change this based on the elbow plot

# Apply K-means with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)  # Fit K-means and get cluster labels

# Add the cluster labels to your DataFrame
X['Cluster'] = cluster_labels  # Add the cluster labels to the original DataFrame

# Use PCA to reduce dimensionality to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)  # Apply PCA to the scaled data

# Create a scatter plot with Plotly
fig = px.scatter(
    x=X_pca[:, 0],  # PCA Component 1
    y=X_pca[:, 1],  # PCA Component 2
    color=cluster_labels,  # Color by cluster
    title="K-means Clustering with PCA Visualization",
    labels={"x": "PCA Component 1", "y": "PCA Component 2"},  # Axis labels
    color_continuous_scale='viridis',  # Color scheme for clusters
)

# Add cluster centers to the plot
cluster_centers_pca = pca.transform(kmeans.cluster_centers_)  # Project cluster centers to 2D
fig.add_trace(
    go.Scatter(
        x=cluster_centers_pca[:, 0],  # PCA Component 1 of cluster centers
        y=cluster_centers_pca[:, 1],  # PCA Component 2 of cluster centers
        mode='markers',
        marker=dict(size=12, color='red'),  # Red markers for cluster centers
        name='Cluster Centers',  # Legend label
    )
)

# Display the Plotly plot
fig.show()
```
